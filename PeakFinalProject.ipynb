```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading in the dataframe 

```python
data = pd.read_csv('amazon.csv')
```

The first five rows in the dataframe.

```python
data.head()
```

Checking for missing data.

```python
missing_data = data.isnull().sum()
missing_data
```

removing the two rating counts 

```python
data = data.dropna()
```

Finding the different item catergories.

```python
catergory_counts = data['category'].value_counts()
catergory_counts
```

```python
catsplit=data['category'].str.split('|',expand=True)
catsplit
```

```python
catsplit=catsplit.rename(columns={0:'category_1',1:'category_2',2:'category_3'})
catsplit
```

```python
data['category_1']= catsplit['category_1']
data['category_2']= catsplit['category_2']
data.drop(columns='category',inplace=True)
data
```

```python
data['category_1']=data['category_1'].str.replace('&',' & ')
data['category_1']=data['category_1'].str.replace('OfficeProducts','Office Products')
data['category_1']=data['category_1'].str.replace('MusicalInstruments','Musical Instruments')
data['category_1']=data['category_1'].str.replace('HomeImprovement','Home Improvement')
catergory_counts=data['category_1'].value_counts()
catergory_counts
```

```python
data['category_2'].value_counts()
```

```python
data['category_2'] = data['category_2'].str.replace('&', ' & ')
data['category_2'] = data['category_2'].str.replace(',', ', ')
data['category_2'] = data['category_2'].str.replace('HomeAppliances', 'Home Appliances')
data['category_2'] = data['category_2'].str.replace('AirQuality', 'Air Quality')
data['category_2'] = data['category_2'].str.replace('WearableTechnology', 'Wearable Technology')
data['category_2'] = data['category_2'].str.replace('NetworkingDevices', 'Networking Devices')
data['category_2'] = data['category_2'].str.replace('OfficePaperProducts', 'Office Paper Products')
data['category_2'] = data['category_2'].str.replace('ExternalDevices', 'External Devices')
data['category_2'] = data['category_2'].str.replace('DataStorage', 'Data Storage')
data['category_2'] = data['category_2'].str.replace('HomeStorage', 'Home Storage')
data['category_2'] = data['category_2'].str.replace('HomeAudio', 'Home Audio')
data['category_2'] = data['category_2'].str.replace('GeneralPurposeBatteries', 'General Purpose Batteries')
data['category_2'] = data['category_2'].str.replace('BatteryChargers', 'Battery Chargers')
data['category_2'] = data['category_2'].str.replace('CraftMaterials', 'Craft Materials')
data['category_2'] = data['category_2'].str.replace('OfficeElectronics', 'Office Electronics')
data['category_2'] = data['category_2'].str.replace('PowerAccessories', 'Power Accessories')
data['category_2'] = data['category_2'].str.replace('CarAccessories', 'Car Accessories')
data['category_2'] = data['category_2'].str.replace('HomeMedicalSupplies', 'Home Medical Supplies')
data['category_2'] = data['category_2'].str.replace('HomeTheater', 'Home Theater')
data['category_2'].value_counts()
```

```python
most_main_items=data['category_1'].value_counts().rename_axis('category_1').reset_index(name='counts')
most_sub_items =data['category_2'].value_counts().head(9).rename_axis('category_2').reset_index(name='counts')
```

```python
colors = sns.color_palette('flare', n_colors=5)
colors_2 = sns.color_palette('pastel', n_colors=5)

fig , ax=plt.subplots(2,1, figsize= (8, 10))
fig.suptitle('Most Amount Product by Category')

sns.barplot(ax=ax[0], data=most_main_items , x='counts',y='category_1', hue='category_1', palette=colors)
sns.barplot(ax=ax[1], data=most_sub_items  , x='counts',y='category_2', hue='category_2', palette=colors_2)

ax[0].set_xlabel('Count')
ax[0].set_ylabel('Product Main Category')

ax[1].set_xlabel('Count',)
ax[1].set_ylabel('Product Sub Category')

ax[0].set_title('Most Products by Main Category')
ax[1].set_title('Most Products by Sub Category')

for container in ax[0].containers:
    ax[0].bar_label(container)

for container in ax[1].containers:
    ax[1].bar_label(container)

plt.show()
```

```python
sorted_rating_data = data.sort_values(by='rating', ascending=False)
sorted_rating_data.head(10)
```

```python
data['rating'] =data['rating'].str.replace('|','4.0').astype('float64')
```

```python
plt.figure(figsize=(10, 6))

# Histogram & Density
sns.histplot(data['rating'], bins=20, kde=True, color='purple', label='Histogram')

# Title and labels
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.legend()
plt.show()
```

The histogram provides a visual representation of the distribution of product ratings, showcasing the frequency of ratings across the dataset. On the other hand, the density plot offers a smoothed representation of the probability density function of the ratings, providing insight into the likelihood of different ratings occurring.

From the plot, we can observe the distribution of ratings among the products. We notice that the distribution is slightly skewed to the left, indicating that a larger proportion of products have higher ratings compared to those with lower ratings. This skewness suggests that most products in the dataset tend to receive positive ratings from users, although there are still variations in the ratings provided.

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='rating', y='category_2', data=data, palette='viridis')
plt.xlabel('Rating')
plt.ylabel('Main Product Category')
plt.title('Boxplot of Ratings by Main Product Category')
plt.show()
```

```python
data['discount_percentage']=data['discount_percentage'].str.replace('%','').astype('float64')
data['discount_percentage']=data['discount_percentage']/100
data['discount_percentage']
```

```python
discount_per_category = data.groupby('category_2')['discount_percentage'].mean()

plt.bar(discount_per_category.index, discount_per_category.values)
plt.xlabel('Sub Category')
plt.ylabel('Mean Discount Percentage')
plt.title('Mean Discount Percentage per Category')
plt.xticks(rotation=90)
plt.show()
```

```python
mean_rating_per_category = data.groupby('category_2')['rating'].mean()

plt.bar(mean_rating_per_category.index, mean_rating_per_category.values)
plt.xlabel('Main Category')
plt.ylabel('Mean Rating')
plt.title('Mean Rating per Category')
plt.xticks(rotation=90)
plt.show()
```
